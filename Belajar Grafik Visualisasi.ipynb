```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('supermarket sales.csv')
df.head()
```

```python
df.tail()
```

```python
fig, ax = plt.subplots() 
ax.plot('Date','gross income', data=df)
plt.show()
```

```python
df.dtypes
```

```python
df['Date']=pd.to_datetime(df.Date)
df['Date']=df['Date'].dt.strftime('%d/%m/%Y')
df['Date']=pd.to_datetime(df.Date)
df.dtypes
```

```python
df.sort_values('Date', inplace=True)
df
```

```python
fig, ax = plt.subplots() 
ax.plot('Date','gross income', data=df)
plt.show()
```

```python
dfgrp= df.groupby('Date', as_index= False)['gross income'].sum()
dfgrp
```

```python
fig, ax = plt.subplots(figsize=(10,8)) 
ax.plot('Date','gross income', data=dfgrp)
plt.show()
```

```python
from datetime import datetime, timedelta
from matplotlib import dates as mpl_dates
fig, ax= plt.subplots(figsize=(10,8))
ax.plot('Date', 'gross income', data=dfgrp, marker='o', linestyle='--', linewidth=1.3)
ax.set(xlabel= 'Period', ylabel='Gross Income', Title= 'Gross Income Trendline')
cdf = mpl_dates.ConciseDateFormatter(ax.xaxis.get_major_locator())
ax.xaxis.set_major_formatter(cdf);
```

```python
dfpgrp= df.groupby('Payment', as_index=False)['Invoice ID'].count()
dfpgrp
```

```python
fig, ax= plt.subplots(figsize=(10,8))
ax.bar('Payment', 'Invoice ID', data=dfpgrp, width= 0.3)
ax.set(xlabel='Payment Type', ylabel='Frequency', Title='Favourite Payment?')
plt.show()
```

```python
dfctgrp= df.groupby('Customer type', as_index=False)['Invoice ID'].count()
dfctgrp
```

```python
fig1, ax1 = plt.subplots()
ax1.pie('Invoice ID', labels='Customer type', data=dfctgrp, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.show()
```
